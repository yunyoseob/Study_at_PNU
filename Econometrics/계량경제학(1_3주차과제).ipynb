# 계량경제학 1주차 과제

계량경제학이란?
--

계량경제학은 경제학이 경제 이론을 바탕으로 중요한 변수들이 상호 어떻게 연계되는지에 관한 이론으로부터 출발한다. 계량경제학은 경제모형을 바탕으로 경제적 결정이나 선택을 하는 데 있어 한 변수의 변화가 다른 변수에 얼마만큼 영향을 미치는가를 알아보는데서 시작한다.

즉, 계량경제학의 정의는 ‘얼마만큼’과 같은 종류의 물음에 답하기 위해 통계학적인 분석도구를 사용하여 경제학, 경영학, 사회과학과 관련된 이론과 자료를 어떻게 이용할 수 있는지에 관해 연구하는 분야이다.

경제학자들은 경제 분석에 있어서 문제과 되고 있는 변수들 간의 관계를 추론하기 위해 경제이론 및 경제논리를 사용한다. 중요한 근간이 되는 모수를 추정하고 예측하기 위해 관련 변수들에 대한 자료를 수집하고 경제적 방법론을 이용한다. 

또한 경영자는 판매액을 예측하기 위하여 계량경제학적인 분석을 담당할 상담역을 채용하려 한다. 이 때, 계량경제학은 계량경제모형을 통해 이러한 문제들을 해결하고자 한다.
계량경제모형은 체계적 부분과 무작위 오차로 이루어진다.

체계적 형태란 경제이론에 따라 개인 또는 기업의 특정한 형태를 예측할 수 없으며 많은 개인 또는 기업의 평균적 또는 체계적 행태를 설명하는 방법이다. 즉, 경제이론과 함수 형태에 관한 가정을 통해 설명할 수 있는 부분을 체계적 형태라고 한다. 이 때 발생하는 무작위적이며 예측할 수 없는 부분을 무작위 오차라고 한다. 즉, 경제활동에 내재된 본질적인 불확실성을 반영하는 오차이다.

계량경제모형을 만들기 위해서는 자료를 구해야 한다. 이 때, 자료는 실험적 자료와 비실험적 자료로 나뉜다. 먼저, 실험적 자료는 경제 관계를 나타내는 알지 못하는 모수에 관해 정보를 얻을 수 있는 방법으로 실험결과를 정리하거나 관찰하는 것이다. 예시로 자연과학 및 농업에서는 통제된 실험을 쉽게 가상해 볼 수 있다.

과학자들은 주요 통제 변수의 값을 정하고 나서 결과를 관찰할 수 있다. 유사한 토지 구획에 특정 품종의 밀을 심고 나서 각 토지 구획에 대해 비료 및 살충제의 양을 상이하게 사용한 후 추수시점에 각 구획에서 생산된 밀의 양을 관찰할 수 있다. 그러나 이런 통제된 실험은 경영이나 사회과학 분야에서 거의 이루어지지는 않는다. 실험적 자료의 주요한 특징은 설명변수들의 값이 반복적으로 이루어지는 실험에서 특정값으로 고정된다는 점이다. 


한편, 비실험적 자료의 예로는 조사자료를 들 수 있다. 예시로 미국 루이지애나 주립대학교에 소재하는 공공정책 연구소는 고객에 대한 전화 및 우편조사를 시행하였다.

전화조사에서 대상 인원은 무작위적으로 뽑아 전화 통화를 하엿다. 질문에 대한 대답은 기록되어 분석이 이루어졌다. 이런 환경에서 모든 변수에 대한 자료는 동시에 수집되며 그 값은 고정되지도 반복되지도 않는다. 이런 자료는 비실험적 자료이다. 수집된 자료는 몇 가지 형태를 가질 수 있다. 자료를 분석할 경우 수집된 형태에 따라 해결해야 할 방법론상의 문제를 갖게 된다. 이렇게 수집된 자료는 시계열 형태, 횡단 형태, 패널자료 형태를 가지며 미시 혹은 거시적 방법으로 통합되어 수집 될 수 있으며, 수집된 자료는 유량, 저량, 양적, 질적일 수 있다.


계량경제학에서 통계적 추론이란 자료의 표본을 분석하여 실제 세계에 대한 것을 ‘추론’하거나 알고자 하는 의미로 사용된다. 통계적 추론을 하는 방법에는 경제모수의 추정, 경제적 결과의 예측, 경제적 가설의 검정이 있다. 계량경제학은 위의 통계적 추론에 관한 모든 면을 포함하고 있다.

# 계량경제학 2주차 과제



미시경제학에서는 생산함수와 총 생산물 곡선으로 생산요소에 대해 알아보고, 거시경제학에서 는 소비함수와 투자함수로 경제의 총투자량이 이자울에 의존하고 총소비가 가처분 소득 수준에 연계시킨다. 계량경제학에서 단순 선형회귀 모형은 회귀모형을 이용하여 한 변수가 변화할 경우 다른 변수가 얼마만큼 변하는지, 한 변수의 값을 아는 경우 다른 변수의 이에 상응하는 값을 예측하거나 예상한다. 이를 토대로 계량경제학에서는 계량경제모형을 설정하는데 계량경제학적 분석을 할 때는 가정을 토대로 분석을 한다. 





단순 선형회귀 모형 기본적 가정
--

단순 선형회귀 모형에 대한 기본적 가정은 다음과 같다. 먼저, 단순 회귀함수를 통해 무작위 표본을 추출할 경우 회귀선에 대해 종모양의 형상이 나타나는데 여기서 평균에 대한 y값의 퍼진 정도를 모든 x에 대해 동일하다고 가정한다. 즉, var(y|x)=σ²인데, 이 조건을 만족시키는 자료를 동분산적이라고 하며, 이 가정을 위배하는 해당 자료를 이분산적이라고 한다.

 다음 가정은 표본이 무작위적이어야 한다. 이는 표본들이 독립적이라는 것을 의미하며, 공분산은 0이거나 cov(yⅰ, yｊ)= 0이라고 가정하는 것이다. 그렇기 때문에 y값들 사이에 선형관계가 없다.

 위의 가정은 y값이 모두 통계적으로 독립적이라고 할 때 더욱 강한 가정이 된다. 또한 변수 x는 확률적이지 않아야 하고 회귀분석의 기본적인 사고는 한 변수 x의 변화가 다른 변수 y에 미치는 영향을 측정하는 것이므로 최소한 2개의 값을 가져야 한다. 마지막 가정으로는 y값은 정규분포되는 것으로 가정한다. 이는 관찰에 기초하고 있으며 가끔은 합리적이지만 선택적인 가정으로 본다. 이를 통해 계량경제모형은 정의되며, 이는 자료를 설명하는 일련의 가정이 된다.

회귀모형에서 일반적으로 종속변수라고 불리는 y의 측면에서 단순 선형회귀 모형에 대한 가정을 설명하는 것이 편리하다. 여기서 조건부 확률밀도함수와는 다르게 x는 일정하며 비확률적이기 때문에 조건부를 의미하는 표기가 필요하지 않다. 이 때 종속변수 y는 독립변수 x와 함께 체계적으로 변화하는 요소와 무작위적 오차항 e로 설명할 수 있다.

오차항
--

 무작위 오차 e는 x이외에 y에 영향을 미치는 모든 요소들을 나타내기 때문에 관찰할 수 없는 특징을 가지고 있다. 이를 통해 x의 각 값에 대해 y값은 y=β1 + β2x + e로 표현할 수 있다.


이 때 오차항 e의 기댓값은 0이 되어야 한다. E(y)= β1 + β2x로 가정하였기 때문이다. 또한 무작위 오차 e의 분산은 var(e) = σ² = var(y)와 같다. 이들은 단지 일정한 상수만큼 차이가 나기 때문이다.

 또한 무작위 오차 e는 서로 독립적이여야 하기 때문에, cov(eⅰ, eｊ)= 0이어야 한다. 그러므로 cov(yⅰ, yｊ)= 0= cov(yⅰ, yｊ)라고 가정할 수 있다. 무작위 오차항 또한 통계적으로 독립적인 경우, 종속변수 y의 값도 통계적으로 독립적이라고 할 경우 이는 더욱 강한 가정이 된다.
 
  마찬가지로 x는 확률적이지 않아야 하고 회귀분석의 기본적인 사고는 한 변수 x의 변화가 다른 변수 y에 미치는 영향을 측정하는 것이므로 최소한 2개의 값을 가져야 한다. 마지막 가정으로는 e값은 정규분포되는 것으로 가정한다.
  
   이는 관찰에 기초하고 있으며 가끔은 합리적이지만 선택적인 가정으로 본다. 이전에 y값이 선택적으로 x의 각 값에 대해 자신의 평균을 중심으로 정규분포가 됐다면, e값은  y값들이 정규분포될 경우 e값은 자신의 평균을 중심으로 정규분포되며 그 역도 성립한다.

최소제곱추정량
--

y 및 x에 대한 표본 관찰값이 주어진 경우 S(β1, β2)= ∑(yi - β1 - β2xi)²
 
 
 ‘제곱의 합’ 함수를 극소화하는 미지의 모수 β1 및 β2의 값을 구하면 최소제곱추정값은 b2 =∑(xi – )(yi - )/ ∑(xi – )², b1=  - b2 이다. (이 때, b1은 함수 S를 β1으로 편미분했을 때 0이 되는 지점, β2는 b2는 함수 S를 β2으로 편미분했을 때 0이 되는 지점을 의미한다.)

이를 전개하기 위해서는 SSE(오차제곱합)을 b1과 b2로 편미분의 값을 0으로 만들면 된다. (SSE를 최소화 하기 위해 함수 S의 표면의 기울기를 0으로 만들기 위해서 SSE를 b1과 b2로 편미분의 값을 0으로 만든다.)

∂S/∂b1 = 2Nb1 – 2∑yi + 2(∑xi)b2 = 0
∂S/∂b2 = 2(∑x²i)b2 – 2∑xiyi + 2(∑xi)b1 = 0


위의 식을 재정리 하면 정규 방정식이 나온다.
① ∑yi = Nb1 +(∑xi)b2
② ∑xiyi = (∑xi)b1+(∑x²i)b2

이를 연립방정식으로 풀면 최소제곱추정값을 구할 수 있다. b2에 대해 풀기 위해서는 ①식에 ∑xi를 곱하고 ②식에 N을 곱하여 ②식에서 ①식을 빼서 좌변에 b2만 남겨놓으면 된다. b1에 대해 풀기 위해서는 ①식의 양쪽 변을 N으로 나누어 재정리해야 한다.

- S(β1, β2)= ∑(yi - β1 - β2xi)²

풀이과정
--

b2 구하는 법
1. ① ∑yi = Nb1 +(∑xi)b2 양변에 ∑xi를 곱한다.
→ ∑xi∑yi = Nb1(∑xi) + b2(∑xi)²

2. ② ∑xiyi = (∑xi)b1+(∑x²i)b2 양변에 N을 곱한다.
→ N∑xiyi = Nb1(∑xi) +Nb2(∑x²i)

3.  ②식에서 ①식을 뺀다.
N∑xiyi - ∑xi∑yi = Nb1(∑xi) +Nb2(∑x²i) -  Nb1(∑xi) - b2(∑xi)²
→ N∑xiyi - ∑xi∑yi = Nb2(∑x²i) - b2(∑xi)²
→ b2{N(∑x²i) - (∑xi)²} = N∑xiyi - ∑xi∑yi

4. 좌변에 b2만 남겨두고 전부 우변으로 정리한다.
b2 = N∑xiyi - ∑xi∑yi/N(∑x²i) - (∑xi)²

b1 구하는 법
1. ① ∑yi = Nb1 +(∑xi)b2 의 양변을 N으로 나누어 정리한다.
→ ∑yi/N = b1 + (∑xi)b2/N

2. 좌변에 b1만 남겨두고 전부 우변으로 정리한다.
→ b1 = 1/N{∑yi – b2(∑xi)}

∴ b1 =  1/N{∑yi – b2(∑xi)}, b2 = N∑xiyi - ∑xi∑yi/N(∑x²i) - (∑xi)²
