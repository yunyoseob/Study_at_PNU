문제) 첨부한 엑셀 파일은 2000~2014년 기간 동안 우리나라 주요 양파 재배지인 전남, 경북, 경남, 제주의 재배면적(ha)과 단위수확량(kg/10a) 자료을 보여준다. 변수명은 “year” (연도), “acr_jn” (전남 재배면적), “acr_kb” (경북 재배면적), “acr_kn” (경남 재배면적), “acr_jj” (제주 재배면적), “yd_jn” (전남 단수), “yd_kb” (경북 단수), “yd_kn” (경남 단수), “yd_jj” (제주 단수)로 구성되어 있다.

```python
df.columns
```

```python
df_jn=df[['year','acr_jn','yd_jn']]
```

```python
df_jn['region']=1
df_jn=df_jn[['region','year','acr_jn','yd_jn']]
df_jn
```

```python
df_kb=df[['year','acr_kb','yd_kb']]
df_kb['region']=2
df_kb=df_kb[['region','year','acr_kb','yd_kb']]
df_kb
```

```python
df_kn=df[['year','acr_kn','yd_kn']]
df_kn['region']=3
df_kn=df_kn[['region','year','acr_kn','yd_kn']]
df_kn
```

```python
df_jj=df[['year','acr_jj','yd_jj']]
df_jj['region']=4
df_jj=df_jj[['region','year','acr_jj','yd_jj']]
df_jj
```

```python
df_1=pd.concat([df_jn, df_kb, df_kn, df_jj], axis=0)
```

```python
df_1=df_1.reset_index()
```

```python
df_1.drop('index', axis=1, inplace=True)
```

```python
df_1.iloc[15:30, 4:6]
```

```python
df_1.iloc[15:30, 2:4]=df_1.iloc[15:30, 4:6]
```

```python
df_1.iloc[30:45, 2:4]=df_1.iloc[30:45, 6:8]
```

```python
df_1.iloc[45:60, 2:4]=df_1.iloc[45:60, 8:10]
```

```python
df_1=df_1.dropna(axis=1)
```

```python
df_1.rename(columns={"acr_jn":"acre", "yd_jn":"yield"}, inplace=True)
```

```python
df_1.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

facet=sns.FacetGrid(df_1, hue="region", aspect=4)
facet.map(sns.kdeplot, 'acre', shade=True)
facet.set(xlim=(0, df_1.acre.max()))
facet.add_legend()
sns.axes_style('dark')
          
plt.show()
```

1: 전남
2: 경북
3: 경남
4: 제주

```python
facet=sns.FacetGrid(df_1, hue="region", aspect=4)
facet.map(sns.kdeplot, 'yield', shade=True)
facet.set(xlim=(0, df_1['yield'].max()))
facet.add_legend()
sns.axes_style('dark')
          
plt.show()
```

```python
df_jn.std(), df_kb.std(), df_kn.std(), df_jj.std()
```

```python
from scipy.stats import skew, kurtosis

skew(df_jn.acr_jn), skew(df_kb.acr_kb), skew(df_kn.acr_kn), skew(df_jj.acr_jj)
```

```python
kurtosis(df_jn.acr_jn), kurtosis(df_kb.acr_kb), kurtosis(df_kn.acr_kn), kurtosis(df_jj.acr_jj)
```

```python
figure, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
figure.set_size_inches(10,10)

sns.distplot(df_jn.acr_jn, ax=ax1)
sns.distplot(df_kb.acr_kb, ax=ax2)
sns.distplot(df_kn.acr_kn, ax=ax3)
sns.distplot(df_jj.acr_jj, ax=ax4)
```

```python
import scipy.stats as stats

print(stats.shapiro(df_jn.acr_jn))
print(stats.shapiro(df_kb.acr_kb))
print(stats.shapiro(df_kn.acr_kn))
print(stats.shapiro(df_jj.acr_jj))
```

귀무가설: 정규분포를 따른다.
대립가설: 정규분포를 따르지 않는다.

p-value가 전부 0.05 이상이므로 귀무가설 채택

```python
skew(df_jn.yd_jn), skew(df_kb.yd_kb), skew(df_kn.yd_kn), skew(df_jj.yd_jj)
```

```python
kurtosis(df_jn.yd_jn), kurtosis(df_kb.yd_kb), kurtosis(df_kn.yd_kn), kurtosis(df_jj.yd_jj)
```

```python
figure, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
figure.set_size_inches(10,10)

sns.distplot(df_jn.yd_jn, ax=ax1)
sns.distplot(df_kb.yd_kb, ax=ax2)
sns.distplot(df_kn.yd_kn, ax=ax3)
sns.distplot(df_jj.yd_jj, ax=ax4)
```
