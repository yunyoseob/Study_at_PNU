# MLE(Maximum Likelihood Estimation)

MLE 정의
--

최대우도법(Maximum Likelihood Estimation)은 모수적인 데이터 밀도 추정 방법으로써 파라미터 $ \theta = (\theta_{1}, ... , \theta_{m})$으로 구성된 어떤 확률밀도함수 $ P(x|\theta)$에서 관측된 표본 데이터 집합을 $x=(x_{1}, x_{2}, ..., x_{n})$이라 할 때, 이 표본들에서 파라미터 $ \theta = (\theta_{1}, ... , \theta_{m})$를 추정하는 방법이다.

**Wikipedia's MLE Definition**

 In statistics, maximum likelihood estimation (MLE) is a method of estimating the parameters of an assumed probability distribution, given some observed data.

# Likelihood function & Maximum Likelihood Estimation





Likelihood: 지금 얻은 데이터가 어떠한 분포로부터 나왔을 가능도를 의미

수치적으로 이 가능도를 계산하기 위해서는 각 데이터 샘플에서 후보 분포에 대한 높이(즉, likelihood 기여도)를 계산해서 다 곱한다.

$$ P(x|\theta) = \prod_{k=1}^n P(x_{k}|\theta) $$

- 전체 표본집합의 결합확률밀도 함수를 likelihood function이라고 한다.

**자연로그를 이용해 log-likelihood function: $ L(\theta|x)$ 만들기**

$$ L(\theta|x) = \log P(x|\theta) = \sum_{i=1}^{n} \log P(x_{i} | \theta) $$

Maximum Likelihood Estimation은 Likelihood의 함수의 최대값을 찾는 방법

- How? 미분을 통해 찾고자하는 파라미터 $\theta$에 대하여 편미분하고 그 값이 0이 되도록 하는 $\theta$를 찾는 과정을 통해 likelihood 함수를 최대화 시켜줄 수 있는 $\theta$를 찾음.

#Definition


**Introduction to Mathematical Statistics (Robert V.Hogg, Joseph Mckean/ Allen T. Craig )**



consider a random variable X whose pdf $f(x;\theta) $ depends on as unknown parameter $\theta$  which is in a set $\Omega$.

For information, suppose that we have a random sample $x_{1}, ..., x_{n}$ are iid random variables with common pdf $f(x;\theta), \theta \in \Omega $.

The parameter $\theta$ is unknown
(we assume that $\theta$ is a scalar)

The basis of our inferential procedures is the likelihood function given by


$$ L(\theta ; x) = \prod_{i=1}^{n} f(x_{i}; \theta) $$

- iid: 각각의 random variable들이 독립 + 동일한 확률분포를 가지는 분포를 뜻한다. 

- 출처: https://dyndy.tistory.com/273 [DY N DY]

- semicolon: 변수와 파라미터를 구분하는 역할

- 출처: http://taewan.kim/post/function_in_semicolon/

Actually, the log of this function is usually more convenient to use and we denote it by

$$ l(\theta) = \log L(\theta) = \sum_{i=1}^{n} \log f(x_{i} ; \theta), \theta \in \Omega $$

our point estimator of $\theta$ is $\hat{\theta}=\hat{\theta} (x_{1} ... x_{n})$, where $\hat{\theta}$ max-imizes the function $L(\theta)$. We call $\hat{\theta}$ the maximum likelihood estimatitor (mle) of $\theta$. 

쉽게 말하면 MLE란, 찾고자 하는 파라미터$\theta$에 대하여 편미분하고 그 값이 0이 되도록 하는 $\theta$를 찾는 과정을 통해 likelihood  함수를 최대화 시켜줄 수 있는 $\theta$를 찾을 수 있다.

$$ \frac{\partial}{\partial \theta} L(\theta|x) = \frac{\partial}{\partial \theta} \log P(x|\theta) = \sum_{i=1}^{n}  \frac{\partial}{\partial \theta} \log P(x_{i}|\theta) = 0 $$

# Assumptions

Let $\theta_{0}$ denote the true value of $\theta$.

(Regularity Conditions: $(R0)-(R2)$ )

**$(R0)$** The cdfs are distinct, that is, $\theta \neq \theta'  \Rightarrow F(x_{i} ; \theta) \neq F(x_{i} ; \theta')  $

- 각각의 누적밀도함수들은 서로 별개이며, 이는 $\theta \neq \theta'  \Rightarrow F(x_{i} ; \theta) \neq F(x_{i} ; \theta')  $를 의미한다.

**$(R1)$** The pdfs have common support for all $\theta$.

-  확률밀도함수들은 모든 $\theta$에 대해서 똑같은 support을 가진다.  $\theta$가 뭐냐에 따라서 support가 영향을 받지 않는다.

**$(R2)$** The point $\theta_{0}$ is an interior point in $\Omega$.

- $\theta_{0}$ 은 $\Omega$(sample space, 표본공간) 바운더리에 있지 않다.

#Maximum Likelihood Estimation


$$ \hat{\theta} = Argmax L(\theta; x)$$

We say that $ \hat{\theta} = \hat{\theta}(x) $, is a maximum likelihood estimator of $\theta$.

$$ L(\hat{\theta} ; x_{1}, ... , x_{n}) = max_{\theta \in \Omega} L(\theta ; x_{1}, ... , x_{n}) $$

log of likelihood and determine its critical value; that is , letting $l(\theta) = \log L(\theta) $, the mle solves the equation.

$$ \frac{\partial l(\theta)}{\partial \theta} = 0 $$

#Theorem 

$ x_{1}, ... x_{n}$ ;  a random sample from $f(x:\theta)$

$\theta_{0}$ : true value of the parameter $\theta$

Under assumptions (R0) and (R1),

$ \lim_{n \rightarrow \infty} P(L(\theta_{0}; x_{1},...x_{n}) > L(\theta ; x_{1}, ... , x_{n})) = 1, \forall \theta \neq \theta_{0} $

#Proof

1. $ L(\theta_{0}; x_{1}, ... , x_{n}) > L(\theta ; x_{1}, ..., x_{n})$

- $ \Leftrightarrow  \prod_{i=1}^{n} \frac{f(x_{i};\theta)}{f(x_{i};\theta_{0})} < 1$

- $ \Leftrightarrow \frac{1}{n} \sum_{i=1}^{n} \log (\frac{f(x_{i};\theta)}{f(x_{i};\theta_{0})}) < 0$

2. By WLLN(대수의 약법칙)

- $ \frac{1}{n} \sum_{i=1}^{n} \log (\frac{f(x_{i};\theta)}{f(x_{i};\theta_{0})}) \xrightarrow{P} E_{\theta_{0}}(\log \frac{f(x_{i};\theta)}{f(x_{i};\theta_{0})})$ 

- $(=^{let} a)$: $E_{\theta_{0}}(\log \frac{f(x_{i};\theta)}{f(x_{i};\theta_{0})})$ 을 a라고 하자.

3. Since $\log x$ is strictly convex, by Jensen's inequality.

- $ a=E_{\theta_{0}}(\log \frac{f(x_{i};\theta)}{f(x_{i};\theta_{0})}) < \log ( E_{\theta_{0}} \frac{f(x_{i};\theta)}{f(x_{i};\theta_{0})}) $

- $ = \log (\int \frac{f(x;\theta)}{f(x;\theta_{0})} f(x;\theta_{0}) dx) $

- $ = \log 1 $

- $ = 0 $

4. since $ a<0$, $\exists$ $\epsilon > 0$ $\exists$ $a+\epsilon<0$

- $P(L(\theta_{0}; x_{1}, ..., x_{n}) > L(\theta ; x_{1}, ... , x_{n} )) $

- $ = P( \frac{1}{n} \sum \log (\frac{f(x;\theta)}{f(x;\theta_{0})} < 0) $

- $= P( \frac{1}{n} \sum \log (\frac{f(x;\theta)}{f(x;\theta_{0})}) < a+ \epsilon)  \xrightarrow{n \rightarrow \infty} 1$ 

> 설명

- $\Leftrightarrow$ : 동치(~는 ~와 모두 참이거나 모두 거짓일 때 성립한다, equivalent)

- log함수는 단조증가함수이므로, L에 로그를 취한 값의 최대 값 역시 우리가 구하려는 $\theta$의 값과 같다. 

- n이 0보다 클 때, 양변에 $\frac{1}{n}$을 곱해줘도 우항은 변하지 않음

- $\xrightarrow{P}$: 화살표 오른쪽 방향으로 수렴한다.
