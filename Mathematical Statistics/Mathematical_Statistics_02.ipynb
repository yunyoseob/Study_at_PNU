# Introduction to Mathematical Statistics

#추정량


모집단의 특성을 알고 싶지만, 모집단 전체를 조사 할 수 없을 때, 우리는 표본을 뽑아 이를 토대로 모집단의 특성을 추정한다.

추정에는 크게 점추정과 구간추정이 있다.

> 점추정

- 추정: 통계량을 통해 모집단의 모수를 추측하는 것

- 점추정: 모집단의 모수를 단일 값으로 추정하는 것

- 추정량: 모수를 추정하는 통계량

- 추정치: 계산된 추정량의 값

> 구간추정

- 구간추정: 모수가 포함될 것으로 기대되는 구간을 추정하는 것이며, 신뢰성 정도를 포함

- 신뢰구간: 구간추정을 통해 얻은 구간 [$\hat{\theta_{L}}$, $\hat{\theta_{U}}$]으로 $\hat{\theta_{L}}$은 신뢰 하한(lower confidence limit), $\hat{\theta_{U}}$는 신뢰 상한(upper confidence limit)을 의미

- 신뢰수준: n번 추출해서 구한 n개의 신뢰구간 중 모수를 포함하는 신뢰구간의 비율

- 오차율: 오차율 $\alpha$는 신뢰구간에 모수가 포함되지 않을 확률

> 좋은 추정량의 조건

- 불편성: 추정량의 기댓값이 모수의 실제값과 같거나 가까울수록 더 좋은 추정량

- 효율성: 모든 불편 추정량 중에서 분산이 작을수록 더 좋은 추정량

- 일치성: 표본의 크기를 크게 할수록 추정량은 모수에 가까워짐

- 충분성: 모수에 대한 정보를 많이 제공할수록 더 좋은 추정량

- 출처: 2021 빅데이터분석기사 필기 (위키북스)

#Unbiased estimator

표본에서 구한 추정량의 기대치가 추정하고자 하는 모수와 같을 때, 그 통계량을 불편추정량이라고 한다. 즉, 편의가 없는 추정량이다. 편의가 없다는 뜻은 추정량이 모수에 대하여 큰 차이를 보이지 않고 어느 한 쪽으로 치우침이 없다는 의미이다.

- 출처: http://www.aistudy.co.kr/math/estimate_lee.htm

우리는 이전에 추정량의 기댓값이 모수의 실제값과 같거나 가까울수록 더 좋은 추정량인 것을 알았다. 즉, $E(\bar{X}) = \mu $, $E(S) = \sigma^2 $이 되어야 좋은 추정량이라는 것이다.

기댓값
--

> 확률변수 X의 확률분포함수가 $f(x)$일 때, X의 기대값 $E(Y)$

- 이산확률변수

$$ E(X) = \sum_x x\cdot f(x) $$

- 연속확률변수

$$ E(X) = \int_{-\infty}^{\infty} x \cdot f(x) dx $$

> 확률변수 X와 Y의 결합확률분포함수가 $f(x,y)$ 일 때

- 이산확률변수

$$ \mu_{x} = E(X) = \sum_x x\cdot f_{X}(x) $$

$$ \mu_{y} = E(Y) = \sum_y y\cdot f_{Y}(y) $$

- 연속확률변수

$$ \mu_{x} = E(X) = \int_{-\infty}^{\infty} x \cdot f_{X}(x) dx $$

$$ \mu_{y} = E(Y) = \int_{-\infty}^{\infty} y \cdot f_{Y}(x) dy $$

- $f_{X}(x), f_{Y}(y)$는 주변확률분포

> 확률변수가 어떤 함수로 주어지는 경우, 확률변수 X와 Y의 결합확률함수가  $f(x,y)$일 때, 확률변수 $g(X,Y)$의 기댓값은

- 이산확률변수

$$ \mu_{g(X,Y)} = E(g(X,Y)) = \sum_{x}\sum_{y} g(x,y)\cdot f(x,y) $$

- 연속확률변수

$$ \mu_{g(X,Y)} = E(g(X,Y)) = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty}  g(x,y) \cdot f(x,y) dxdy $$

표본평균$\cdot$표본분산$\cdot$표본표준편차
--

- 표본평균: $\bar{X} = \frac{1}{n}(X_{1}+X_{2}+X_{3}+\cdots+ X_{n})$

- 표본분산: $ S^2 = \frac{1}{n-1} \sum_{i=1}^{n}(X_{i} - \bar{X} )^2 $

- 표본표준편차: $ S=\sqrt{S^2}$


- 출처: https://koreanfoodie.me/443


> Q. 왜 표본분산은 $n-1$로 나누죠?

> A.

$ E(s^2) = E(\frac{1}{n} \sum_{i=1}^{n} (X_{i}- \bar{X})^2) $

$ = \frac{1}{n} E(\sum_{i=1}^{n} (X_{i}^2 - 2X_{i}\bar{X} + \bar{X}^2)$

$ = \frac{1}{n} E(\sum_{i=1}^{n} X_{i}^2 - 2\bar{X} \sum_{i=1}^{n} X_{i} + \sum_{i=1}^{n} \bar{X}^2 )$

$ = \frac{1}{n} E(\sum_{i=1}^{n} X_{i}^2 - 2n\bar{X}^2 + n\bar{X}^2) $

$ = \frac{1}{n} \sum_{i=1}^{n} E(X_{i}^{2}) - E(\bar{X}^2) $

$ = \frac{1}{n} \sum_{i=1}^{n} (\mu^2 + \sigma^2 ) - (\mu^2 + \frac{\sigma^{2}}{n} )$

= $\frac{n-1}{n} \sigma^2 $




계산해보면 $E(s^2) = \sigma^2 $를 만족하지 않기 때문에, 표본분산은 분편성을 만족하는 불편 추정량이 아니다. 따라서 표본분산이 좋은 추정량이 되게 하기 위해 분모 $n$을 $n-1$으로 바꾸어주는 것이다.

#Uniformly Minimum Variance Unbiased Estimation

**Lehmann-Scheffe 정리**

Uniformly Minimum Variance Unbiased Estimation(최소불편추정량)은 $Y=u(X_1, X_2, \cdots, X_n)$을 $\theta$에 대한 완비충분통계량(CSS)이라고 할 때, 임의의 불편추정량 $\theta$에 대해 $ \psi(Y) = E[\hat{\theta}|Y]$는 $theta$에 대한 최소분산분편추정량이며,  이는 유일(unique) 하다.

CSS(Complete and Sufficient Statistic) 함수로서 $(\psi(Y))$ 기댓값이 특정 모수($\theta$ 또는 $\theta$의 함수)가 될 때, $psi(Y)$가 바로 그 모수에 대한 UMVUE가 된다.

- 출처: https://syj9700.tistory.com/5

> Completeness and Uniqueness

$ z \sim h(z;\theta), \theta \in \Omega $

$ u(\cdot)$ ; free of $\theta$

The family $\{h(z;\theta) ; \theta \in \Omega \} $ of pdf's is complete iff

$ E(u(z))=0 $ $\forall 0 $ implies $u(z)=0$

except on a set of prob. zero $\forall 0 \in \Omega $

- iff: 정의는

- $ u(\cdot)$: 어떤 함수

- except on a set of prob. zero $\forall 0 \in \Omega $ : 확률 1로써 0함수다.(값이 변하지 않기 때문에)




즉, 쉽게 말해서 unbiased estimator중 Best를 찾는다.

> Complete Statistic

$ X_1, \cdots , X_n$; r.s from $f(x;\theta), \theta \in \Omega$

$ Y= u(X_1, \cdots, X_n)$ ; a statistic with pdf $g(y;\theta)$

$\rightarrow$ Y is said to be a complete statistic

if the family $\{g(y;\theta) ; \theta \in \Omega \}$ is complete

> Uniformly Minimum Variance Unbiased Estimation

$ X_1, \cdots , X_n$; r.s from $f(x;\theta), \theta \in \Omega$

Let $Y=u(X_1, \cdots, X_n)$ be an estimator of $T(\theta)$ (:Top of theta) which statisfies

1) $E(Y)=Y(\theta), \forall \theta \in \Omega \$ (unbiasedness)

2) $Var(Y) \leq Var(T), \forall \theta \in \Omega $ for any unbiased estimator $T(\theta)$

Then Y is called Uniformly Minimum Variance Unbiased Estimation


- Y, T 둘 다 unbiased estimate

- $Var(Y) \leq Var(T)$ : 어떠한 T를 가져와도 성립이 된다. Variance는 Mean Squared Error(MSE)가 됨, Variance가 낮을 수록 MSE가 떨어져 좋은 estimate 가 된다.

- Uniformly: MSE, Variacnce는 $\theta$의 파라미터 함수

- UMVUE: 가능한 모든 $\theta$에 대해서 minimum variance를 갖는 unbiased estimator

- **주의**

-  Best Unibiased Estimator 이면서 Best라고 착각하는 경우가 많은데, Unbiased Estimator들 중에서 Best라는 뜻이다.

CSS(Complete and Sufficient Statistic)를 말로 다시 설명해보자
--

어떤 분포를 따르는 확률표본 $X_1, X_2, \cdots, X_n$이 존재한다고 할 때, 모수 $\theta$를 추정하기 위해 n개의 표본을 이용할 수도 있지만, n개의 표본을 더한 하나의 값(충분통계량)만으로도 모수를 추정할 수 있다.

n개의 표본이 가지고 있는 $\theta$에 대한 모든 정보와 동일한 정보력(sufficiency)을 가지며,
단순한 형태(minimal type)을 띄는 것으로 설명할 수 있다.

Sufficient Statistic
--

$X_1, X_2, \cdots, X_n \sim f(x;\theta)$로부터의 확률표본이라고 하자. 어떤 통계량 $Y=u(X_1, X_2, \cdots, X_n)$에 대해, $Y=y$일 때, $X|Y=y$의 분포가 $\theta$에 의존하지 않을 때, Y를 $\theta$에 대한 충분통계량(Sufficient Statistic or S.S)이라고 한다. 

Completeness
--

통계량 Y가 분포모임의 한 원소를 확률밀도함수로 가진다고 하자. 모든 $\theta \in \Omega$에 대해 $E_{\theta}[\psi(Y)] = 0$를 만족하는 $\psi$가 $\psi(y)=0$뿐이라면, 위 분포족이 완비성을 가진다고 하고, Y를 완비통계량(Complete Statistic)이라고 한다.

- $\psi$: $\theta$에 무관한 함수

Complete and Sufficient Statistic
--

통계량 Y가 $\theta$에 대한 충분통계량이며 동시에 완비통계량일 때, 통계량 Y를 $\theta$에 대한 완비충분통계량이라고 한다.

# Cramer-Rao 부등식

임의의 불편추정량이 가질 수 있는 분산의 하한(lower bound)를 제공한다. 최소분산을 갖는 불편추정량은 유일하므로, 주어진 불편추정량의 분산이 Cramer-Rao 하한과 일치하면 그 추정량이 유일한 UMVUE(최소분산불편추정량)이 된다.

# Cramer-Rao Lower bound and efficiency

likelihood Assumptions



Let $\theta_{0}$ denote the true value of $\theta$.

(Regularity Conditions: $(R0)-(R2)$ )

**$(R0)$** The cdfs are distinct, that is, $\theta \neq \theta'  \Rightarrow F(x_{i} ; \theta) \neq F(x_{i} ; \theta')  $

- 각각의 누적밀도함수들은 서로 별개이며, 이는 $\theta \neq \theta'  \Rightarrow F(x_{i} ; \theta) \neq F(x_{i} ; \theta')  $를 의미한다.

**$(R1)$** The pdfs have common support for all $\theta$.

-  확률밀도함수들은 모든 $\theta$에 대해서 똑같은 support을 가진다.  $\theta$가 뭐냐에 따라서 support가 영향을 받지 않는다.

**$(R2)$** The point $\theta_{0}$ is an interior point in $\Omega$.

- $\theta_{0}$ 은 $\Omega$(sample space, 표본공간) 바운더리에 있지 않다.

+++++++

$x_{i} \sim iid f(x;\theta), i=1, \cdots, n, \theta \in \Omega. $

$\Omega$ is an open interval.


**(R3)** $f(x;\theta)$ is twice differentiable w.r.t $\theta$

- $f(x;\theta)$는 $\theta$에 대해서 두 번 미분 가능하다.

**(R4)** Let $u(x)$ be such that $\int u(x) f(x;\theta) dx < \infty $ 

- $\int u(x) f(x;\theta) dx = E(u(X))$

Then the integral $\int u(x)f(x;\theta)dx$ is twice differentiable and the derivatives can be obtained by differentiating under the integral sign.

- 이 경우에  $\int u(x)f(x;\theta)dx$은 두 번 미분 가능하고 첫 번째 미분과 두 번째 미분이 존재하는데, 한 번 미분, 두 번 미분한 것을 적분한 것과 x에 대해서 적분하고 나서 $\theta$에 대해서 한 번 미분, 두 번 미분한 것을 적분한 것과 같다. 


Since $ 1= \int f(x;\theta) dx, $, under(R0) ~ (R4)

$ 0 = \frac{\partial}{\partial \theta} \int f(x;\theta) dx $

$ = \int \frac{\partial}{\partial \theta} f(x;\theta) dx $

$ = \int \frac{\frac{\partial}{\partial \theta} f(x;\theta)}{f(x;\theta} \cdot f(x;\theta) dx $

$ = \int \frac{\partial}{\partial \theta} \log f(x;\theta) \cdot f(x;\theta) dx $

$ = E(\frac{\partial}{\partial \theta} \log f(x; \theta)) $

and

$ 0 = \frac{\partial^2}{\partial \theta^2} \int f(x;\theta) dx$

$ = \int \frac{\partial}{\partial \theta} ( \frac{\partial}{\partial \theta} \log f(x;\theta) f(x;\theta)) dx $

$  = \int \frac{\partial^2}{\partial \theta^2} \log f(x;\theta) \cdot f(x;\theta))dx + \int  \frac{\partial}{\partial \theta} \log f(x;\theta)^2 f(x;\theta))dx$



Fisher Information
--

if it exists

$ I(\theta) = E_{\theta}((\frac{\partial}{\partial \theta} \log f(x;\theta))^2) $

Under (R0) ~ (R4)

$ I(\theta) = - E_{\theta}(\frac{\partial^2}{\partial \theta^2} \log f(x;\theta)) $

- The function $\frac{\partial}{\partial \theta} \log f(x;\theta)$ is called score function

Cramer-Rao Lower bound
--

$ X_{1}, \cdots , X_{n} \sim iid f(x;\theta), \theta \in \Omega $

Assume the regularity conditions (R0)~(R4) hold.

Let $Y=u(X_{1}, \cdots, X_{n})$ be a statistic with mean

$ E(Y) = E(u(X_{1}, \cdots, X_n)  let = K(\theta) $

Then $Var(Y) \geq \frac{(k'(\theta))^2}{n\cdot I(\theta)} $
